```python
# Loan Approval Expert System
```

```python
def approve_loan(income, credit_score, existing_loan):

    if income >= 50000 and credit_score >= 700 and not existing_loan:
        return "Loan Approved"

    elif income >= 30000 and credit_score >= 650:
        return "Further Review Required"

    else:
        return "Loan Rejected"
```

```python
# Facts
```

```python
income = 70000
credit_score = 750
existing_loan = False
```

```python
# Inference Engine
```

```python
decision = approve_loan(
    income,
    credit_score,
    existing_loan
)
```

```python
print("Decision:", decision)
```
